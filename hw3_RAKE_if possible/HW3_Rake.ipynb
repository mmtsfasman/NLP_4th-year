**1.3.** –ü—Ä–µ–¥–ª–æ–∂–∏—Ç—å —É–ª—É—á—à–µ–Ω–∏–µ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –∏ –∏–º–ø–ª–µ–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –µ–≥–æ<br />

–ú–Ω–µ –∫–∞–∂–µ—Ç—Å—è, —á—Ç–æ –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏ –ø–æ—á–∏—Å—Ç–∏—Ç—å —Ç–µ–∫—Å—Ç –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å.

```python
#clean and lemmatize text
import re
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
all_words = re.split(r'\W+', text)
all_words_clean = []
for w in all_words:
    if re.search('[A-Za-z]', w) == None:
        pass
    else:
        w = w.lower()
        new_w = wnl.lemmatize(w)
        if new_w == w:
            new_w = wnl.lemmatize(w, 'v')
        all_words_clean.append(new_w)
clean_text = ' '.join(all_words_clean)
```

**(4) —Ç–æ–ø 15 , –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —É–ª—É—á—à–µ–Ω–Ω—ã–º Rake**

**–ù–∞ —Ä—É—Å—Å–∫–æ–º –ø–æ–ª—É—á–∏–ª–æ—Å—å –∫–∞–∫-—Ç–æ —Å–æ–≤—Å–µ–º –Ω–µ –æ—á–µ–Ω—å...** –ù–∞–≤–µ—Ä–Ω–æ–µ, —ç—Ç–æ —Å–≤—è–∑–∞–Ω–æ —Å —Ç–µ–º, —á—Ç–æ —Ä—ç–π–∫ –Ω–µ –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä—É–µ—Ç —Å–ª–æ–≤–∞, –∞ –≤ —Ä—É—Å—Å–∫–æ–º –Ω–∞–º–Ω–æ–≥–æ –±–æ–≥–∞—á–µ –º–æ—Ä—Ñ–æ–ª–æ–≥–∏—è.

```python
#–ø–æ—á–∏—Å—Ç–∏–º —Ç–µ–∫—Å—Ç
import re
from pymystem3 import Mystem
m = Mystem()
all_words = m.lemmatize(text)
all_words_clean = []
for w in all_words:
    if re.search('\w', w) == None:
        pass
    else:
        all_words_clean.append(w)
clean_text = ' '.join(all_words_clean)
```
