```python
#функция обработки текста
def text_to_words(filename, stop_words):
    with open(filename) as f:
        text = f.read()
        all_words = m.lemmatize(text)

    all_words_clean = []
    for w in all_words:
        if w in stop_words or re.search('[А-Яа-я]', w) == None:
            pass
        else:
            all_words_clean.append(w)
    
    return all_words_clean
```

```python
def weirdness(Nd, Nl, cd, cl):
    '''Nd = total Spec
        Nl = total Ref
        cd = frequency word Spec
        cl = frequency word Ref
    '''
    return ((cd/Nd)/(cl/Nl))
```

```python
def loglikelihood(a, b, c, d):
    ''' 
    a = frequency word Spec
    b = frequency word Ref
    c = total Spec
    d = total Ref   
    '''
    E1 = (c*(a+b))/(c+d)
    E2 = (d*(a+b))/(c+d)
    ln1 = np.log(a/E1)
    ln2 = np.log(b/E2)
    return (2*((a*ln1) + (b*ln2)))
```

```python
def word_frequency(text):            
    #функция, подсчитывающая частоту каждого слова в корпусе text
    frequency = {}
    for word in text:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1 
    return frequency
```

```python
Ref = text_to_words('news.txt', stop_words)
Spec = text_to_words('reviews_test.txt', stop_words)
frequencyRef = word_frequency(Ref)
frequencySpec = word_frequency(Spec)

loglikelihood_total = {}
weirdness_total = {}
for word in set(Spec):
    if word in frequencyRef:
        loglikelihood_total[word] = loglikelihood(frequencySpec[word], frequencyRef[word], len(Spec), len(Ref))
        weirdness_total[word] = weirdness(len(Spec), len(Ref), frequencySpec[word], frequencyRef[word])
    
    
```

```python
#сортировка по возрастанию лог лайклихуд и weirdness
sorted_by_loglikelihood = sorted(loglikelihood_total.items(), key=lambda x: float(x[1]), reverse = True)
sorted_by_weirdness = sorted(weirdness_total.items(), key=lambda x: float(x[1]), reverse = True)
```

```python
#костыль, преобразующий сортированные списки обратно в словарь
weirdness_sorted_dict = {}
for el in sorted_by_weirdness:
    weirdness_sorted_dict[el[0]] = el[1]
    

loglikelihood_sorted_dict = {}
for el in sorted_by_loglikelihood:
    loglikelihood_sorted_dict[el[0]] = el[1]    
```

```python
#feel up the dataframe
data = {}
words = []
word_type = []
countspec = []
countref = []
loglikelihood = []
rang = []
weirdness = []
rang_weirdness = []

for n, tup in enumerate(sorted_by_loglikelihood):
    if n <= 10:
        word = tup[0]
        words.append(word)
        word_type.append('тематическое')
        countspec.append(frequencySpec[word]/len(Spec))
        countref.append(frequencyRef[word]/len(Ref))
        loglikelihood.append(loglikelihood_total[word])
        rang.append(n)
        weirdness.append(weirdness_total[word])
        rang_weirdness.append(sorted_by_weirdness.index((word, weirdness_total[word])))
        
    else:
        break
        
        

```

```python
#feel up the dataframe
sorted_by_loglikelihood_before = sorted_by_loglikelihood
sorted_by_loglikelihood = sorted(loglikelihood_total.items(), key=lambda x: float(x[1]))


for n, tup in enumerate(sorted_by_loglikelihood):
    if n <= 10:
        word = tup[0]
        words.append(word)
        word_type.append('общеупотребительное')
        countspec.append(frequencySpec[word]/len(Spec))
        countref.append(frequencyRef[word]/len(Ref))
        loglikelihood.append(loglikelihood_total[word])
        rang.append(sorted_by_loglikelihood_before.index((word, loglikelihood_total[word])))
        weirdness.append(weirdness_total[word])
        rang_weirdness.append(sorted_by_weirdness.index((word, weirdness_total[word])))
    else:
        break
        
        
```

```python
#draw the table
data = collections.OrderedDict(data)
data['word'] = words
data['Тип'] = word_type
data['CountSpecC'] = countspec
data['CountRefC'] = countref
data['LogLikelihood'] = loglikelihood
data['Ранг loglikelihood'] = rang
data['Weirdness'] = weirdness
data['Ранг weirdness'] = rang_weirdness

pandas.DataFrame(data = data)
```
