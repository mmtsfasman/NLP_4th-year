**1.3.** Предложить улучшение алгоритма и имплементировать его<br />

Мне кажется, что лемматизировать и почистить текст может помочь.

```python
#clean and lemmatize text
import re
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
all_words = re.split(r'\W+', text)
all_words_clean = []
for w in all_words:
    if re.search('[A-Za-z]', w) == None:
        pass
    else:
        w = w.lower()
        new_w = wnl.lemmatize(w)
        if new_w == w:
            new_w = wnl.lemmatize(w, 'v')
        all_words_clean.append(new_w)
clean_text = ' '.join(all_words_clean)
```

**(4) топ 15 , полученных улучшенным Rake**

```python
#recreate RAKE list on clean text

keywords2 = rake_object.run(clean_text, maxWords = 3)

results2 = pd.DataFrame.from_records(keywords2[:15], columns=['keywords', 'score'])
results2
```

**Сравниваем улучшенный и старый рейк**

```python
pd.concat([results1, results2], axis = 1)
```

**Похоже, это помогло.**

Появились следующие адекватно значимые словосочетания, которые до этого не выделял рейк:
* computer vision algorithm
* hydro acoustic system
* modern fish farm
* feed fish visually
* artificial intelligence
* automate fish monitor	

**Если до этого только 7 слов из 15 были хоть как-то значимы, то в улучшенном рейке 11 соответствуют действительности**

**На русском получилось как-то совсем не очень...** Наверное, это связано с тем, что рэйк не лемматизирует слова, а в русском намного богаче морфология.

```python
#почистим текст
import re
from pymystem3 import Mystem
m = Mystem()
all_words = m.lemmatize(text)
all_words_clean = []
for w in all_words:
    if re.search('\w', w) == None:
        pass
    else:
        all_words_clean.append(w)
clean_text = ' '.join(all_words_clean)
```

```python
#создадим новый список на чистом лемматизированном тексте
rake_object = rake.Rake("SmartStoplist_ru.txt")
sample_file = open("medical_text.txt", 'r')
text = sample_file.read()
keywords = rake_object.run(clean_text, maxWords = 3)
#посмотрим на табличку
pd.DataFrame.from_records(keywords[:15], columns=['keywords', 'score'])
```
