```python
import RAKE as rake
import operator
import io
import pandas as pd
```

```python
#create list of keywords using RAKE 
rake_object = rake.Rake("SmartStoplist.txt")
sample_file = open("Fish_farming.txt", 'r')
text = sample_file.read()
keywords = rake_object.run(text, maxWords = 3)
```

* <span style="color:#00868b">**top-15 keywords returned by RAKE**</span>

```python
#create table of keywords and their scores
pd.DataFrame.from_records(keywords[:15], columns=['keywords', 'score'])
```

It seems like following expressions are not relevant here:
* 7m-10m krone
* mr.shang hopes ))
* salmon sloshing loudly
* salmon prices soaring
* humans simply overseeing	
* believes mr sovegjarto

It may be due to untokenized text. Lets try to tokenize it first.

```python
#clean and lemmatize text
import re
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
all_words = re.split(r'\W+', text)
all_words_clean = []
for w in all_words:
    if re.search('[A-Za-z]', w) == None:
        pass
    else:
        w = w.lower()
        new_w = wnl.lemmatize(w)
        if new_w == w:
            new_w = wnl.lemmatize(w, 'v')
        all_words_clean.append(new_w)
clean_text = ' '.join(all_words_clean)
```

```python
#recreate RAKE list on clean text
rake_object = rake.Rake("SmartStoplist.txt")
sample_file = open("Fish_farming.txt", 'r')
text = sample_file.read()
keywords = rake_object.run(clean_text, maxWords = 3)

pd.DataFrame.from_records(keywords[:15], columns=['keywords', 'score'])
```

**На русском получилось как-то совсем не очень...** Наверное, это связано с тем, что рэйк не лемматизирует слова, а в русском намного богаче морфология.

```python
#почистим текст
import re
from pymystem3 import Mystem
m = Mystem()
all_words = m.lemmatize(text)
all_words_clean = []
for w in all_words:
    if re.search('[А-Яа-я]', w) == None:
        pass
    else:
        all_words_clean.append(w)
clean_text = ' '.join(all_words_clean)
```

```python
#создадим новый список на чистом лемматизированном тексте
rake_object = rake.Rake("SmartStoplist_ru.txt")
sample_file = open("medical_text.txt", 'r')
text = sample_file.read()
keywords = rake_object.run(clean_text, maxWords = 2)
#посмотрим на табличку
pd.DataFrame.from_records(keywords, columns=['keywords', 'score'])
```

**По-моему тут лемматизация помогла! Ура! Хотя тест по-прежнему совсем не точный, но в списке есть и словосочетания специфичные для текста.**
